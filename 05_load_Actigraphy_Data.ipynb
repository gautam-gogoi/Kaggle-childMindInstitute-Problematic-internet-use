```python
import numpy as np 
import pandas as pd 
```

```python
def overall_vars(var_name, prefix, df_vars, df):
    summary = pd.DataFrame(df[var_name].describe()).transpose()
    summary.columns = prefix + "_" + summary.columns 
    summary.reset_index(drop=True, inplace=True)
    return (pd.concat([df_vars, summary], axis=1))
```

```python
def create_vars(df, _id):
    data = {'id': [_id] }
    df_output = pd.DataFrame(data)
    vars_for_overall = ['X', 'Y', 'Z', 'enmo', 'anglez']
    for i in vars_for_overall:
        #print(i)
        df_output = overall_vars(i, i, df_output, df)

    df["not_worn_flag_custom"] = np.round(df["non-wear_flag"])
    #df[df["not_worn_flag_custom"]==0].info()
    vars_for_overall = ['X', 'Y', 'Z', 'enmo', 'anglez']
    for i in vars_for_overall:
        prefix=i+"_wear"
        #print(prefix)
        df_output = overall_vars(i, prefix, df_output, df[df["not_worn_flag_custom"]==0])

    for i in vars_for_overall:
        prefix=i+"_weekends"
        #print(prefix)
        df_output = overall_vars(i, prefix, df_output, df[df["weekday"]>=6])

    for i in vars_for_overall:
        prefix=i+"_night"
        #print(prefix)
        df_output = overall_vars(i, prefix, df_output, df[df["time_of_day"]<=2.158000e+13])
    
    return(df_output)
```
